```python
# Gerekli kütüphaneyi yüklüyoruz
!pip install transformers sentencepiece
```

```python
model_adi = "bert-base-multilingual-cased"
```

```python
from transformers import AutoTokenizer

# Sunumdaki BERT gibi bir Encoder modelinin tokenizer'ını kullanalım
tokenizer = AutoTokenizer.from_pretrained(model_adi)
```

```python
cumle = "Türkiye Sevakin Adasını 99 yıllığına Sudan'dan aldı."
```

## Tokenları görme
Not: `Bert` tokenizer'ı `encode` fonksiyonunda başlangıca [CLS] token'ı (ID: 101), sona [SEP] token'ı (ID: 102) ekler.

```python
# Token'lara ayırma
tokenlar = tokenizer.tokenize(cumle)
print(f"Cümle: {cumle}")
print(f"Token'lar: {tokenlar}")

# Token'ları ID'lere (sayılara) çevirme
input_idler = tokenizer.encode(cumle)
print(f"Input ID'leri: {input_idler}")
```

```python
from transformers import pipeline
```

```python
# Türkçe duygu analizi için eğitilmiş bir model kullanıyoruz
encoder_model_adi = "savasy/bert-base-turkish-sentiment-cased"
```

```python
sentiment_pipeline = pipeline("sentiment-analysis", model=encoder_model_adi)
```

```python
olumlu_cumle = "Ürünün fiyatı oldukça makul ve ürün çok güzeldi."
olumsuz_cumle = "Yemekler çok kötüydü ve servis berbattı."
```

```python
print(f"Cümle: {olumlu_cumle}")
print("Sonuç:", sentiment_pipeline(olumlu_cumle))
print("-" * 30)
print(f"Cümle: {olumsuz_cumle}")
print("Sonuç:", sentiment_pipeline(olumsuz_cumle))
```

# Decoder Mimarisi

```python
decoder_model_adi = "ytu-ce-cosmos/turkish-gpt2"
```

```python
# Türkçe metin üretimi için eğitilmiş bir GPT-2 modeli
generator = pipeline('text-generation', model = decoder_model_adi)
```

```python
prompt = "Türkiye S400 hava savunma sistemini"
```

```python
# Daha belirleyici (düşük temperature)
print("--- Temperature=0.7 (Daha Tahmin Edilebilir) ---")
result1 = generator(
    prompt,
    max_new_tokens=50,  # max_length yerine max_new_tokens kullan
    num_return_sequences=1,
    temperature=0.7,
    do_sample=True,
    repetition_penalty=1.2,  # Tekrarları cezalandır
    truncation=True,  # Truncation uyarısını önler
    pad_token_id=50256  # GPT-2 için eos_token_id
)
print(result1[0]['generated_text'])

print("\n--- Temperature=1.5 (Daha Özgün/Rastgele) ---")
result2 = generator(
    prompt,
    max_new_tokens=50,
    num_return_sequences=1,
    temperature=1.5,
    do_sample=True,
    repetition_penalty=1.3,  # Yüksek temperature'da daha fazla ceza
    truncation=True,
    pad_token_id=50256
)
print(result2[0]['generated_text'])
```

# Encoder-Decoder Mimarisi

```python
from transformers import pipeline
```

```python
encoder_decoder_model_adi = "Helsinki-NLP/opus-mt-ar-en"
```

```python
# Fransızca'dan İngilizce'ye çeviri için eğitilmiş bir model
translator = pipeline("translation", model=encoder_decoder_model_adi)
```

```python
arapca_cumle = "تركيا بلد قوي جدًا."

print(f"Arapça Cümle: {arapca_cumle}")
print("Tercüme Sonucu:", translator(arapca_cumle)[0]['translation_text'])
```
